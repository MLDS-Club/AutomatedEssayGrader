```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_feather('~/Documents/GitHub/AutomatedEssayGrader/essaygrader/data/prelimData.feather')
df.columns
```

```python
essay_set_1 = df.loc[df['essay_set'] == 1]
essay_set_2 = df.loc[df['essay_set'] == 2]
essay_set_3 = df.loc[df['essay_set'] == 3]
essay_set_4 = df.loc[df['essay_set'] == 4]
essay_set_5 = df.loc[df['essay_set'] == 5]
essay_set_6 = df.loc[df['essay_set'] == 6]
essay_set_7 = df.loc[df['essay_set'] == 7]
essay_set_8 = df.loc[df['essay_set'] == 8]

scores1 = essay_set_1['normalized_score'].values
scores2 = essay_set_2['normalized_score'].values
scores3 = essay_set_3['normalized_score'].values
scores4 = essay_set_4['normalized_score'].values
scores5 = essay_set_5['normalized_score'].values
scores6 = essay_set_6['normalized_score'].values
scores7 = essay_set_7['normalized_score'].values
scores8 = essay_set_8['normalized_score'].values
```

```python
# Scatter plots for average word length
fig, ax = plt.subplots(8, figsize=(20, 45))
ax[0].scatter(essay_set_1['Average Word Length'], scores1)
ax[1].scatter(essay_set_2['Average Word Length'], scores2)
ax[2].scatter(essay_set_3['Average Word Length'], scores3)
ax[3].scatter(essay_set_4['Average Word Length'], scores4)
ax[4].scatter(essay_set_5['Average Word Length'], scores5)
ax[5].scatter(essay_set_6['Average Word Length'], scores6)
ax[6].scatter(essay_set_7['Average Word Length'], scores7)
ax[7].scatter(essay_set_8['Average Word Length'], scores8)

# title the graph
for i in range(8):
  ax[i].set_title('Scores vs Average Word Length for Essay Set ' + str(i+1))

# label the x and y axes
for i in range(8):
  ax[i].set_xlabel('Average Word Length')
  ax[i].set_ylabel('Scores')
# fig.savefig('images/averageWordLength.svg')
```
